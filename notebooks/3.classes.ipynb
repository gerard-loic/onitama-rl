# Classes "générales"

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| Board | board.py | Représentation d'un état du jeu Onitama |
| Card | card.py | Représentation d'une carte de mouvement |
| Game | game.py | Joue une partie contre P1 et P2 |
| GameSession | game.py | Joue un nombre N de parties entre P1 et P2 |
| Heuristic | heuristic.py | Permet l'évaluation heuristique (différentes méthoides) d'un état |
| Trainer | trainer.py | ? |
| Player | players.py | Classe parente de tous les joueurs |
| HumanPlayer | players.py | Joueur humain (en console) |
| DataTrainer | trainer.py | Classe parente de tous les data trainer |
| RegularDataTrainer | trainer.py | Classe permettant la gestion des parties visant à générer des données d'entraînement |

# Joueurs basés sur des techniques de RL classiques

| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| RandomPlayer | players.py | Joueur aléatoire |
| HeuristicPlayer | players.py | Joueur basé sur des règles d'heuristique (différentes stratégies) |
| LookAheadHeuristicPlayer | players.py | Joueur basé sur des règles d'heuristique (différentes stratégies) + algorithme minimax sur N niveaux |
| MCTSPlayer | players.py | Joueur basé sur l'algorithme Monte Carle Tree Search |
| MCTSHeuristicPlayer | players.py | Joueur basé sur un mélange entre l'algorithme Monte Carlo Tree Search et des règles d'heuristiques, pour diminuer le cout computationnel |

# Joueurs basés sur des réseaux de neurones
| Classe | Fichier | Usage |
| ------ | ------- | ----- |
| CNNPlayer_v1 | dl_players_v1.py | Joueur basé sur un réseau comprenant 1 couche CNN et 5 blocs résiduels + 2 têtes (policy/value) |
| CNNPlayer_v2 | dl_players_v2.py | Joueur basé sur un réseau basé sur la même architecture que v1 mais utilisant l'optimiseur AdamW et implémentant la possibilité d'utiliser une heuristique en addition |
| CNNPlayer_v3 | dl_players_v3.py | Joueur basé sur un réseau comprenant 1 couche CNN simplifiée et 2 blocs résiduels + 2 têtes (policy/value) |
