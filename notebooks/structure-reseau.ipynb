Structure du reseau (de haut en bas) :

Input (5, 5, 10) - les 10 canaux detailles (4 plans de pieces, 5 plans de cartes, 1 plan d'identite joueur)
Convolution initiale - Conv2D(128, 3x3) + BatchNorm + ReLU pour projeter les 10 canaux bruts en 128 features
5 Blocs residuels - chacun avec 2 convolutions, BatchNorm, et une skip connection
Bifurcation vers les deux tetes :
Tete Policy (gauche) : Conv1x1(32) -> Conv1x1(52) -> Reshape(1300) - produit les logits d'action
Tete Value (droite) : Conv1x1(4) -> Flatten -> Dense(64) -> Dense(1, tanh) - evalue la position [-1, +1]
Section Entrainement - les loss functions (CrossEntropy poids 1.0 / MSE poids 0.5)
Justifications architecturales annotees :

Inspiration AlphaZero : tronc residuel partage + double tete, architecture prouvee pour les jeux de plateau
Adaptation au plateau 5x5 : 128 filtres et 5 blocs (vs 256/19-39 pour le Go 19x19), reseau plus leger adapte a l'espace d'etats plus restreint d'Onitama
Regularisation : BatchNorm + Dropout 0.3 (absent dans AlphaZero original mais necessaire ici avec un dataset plus petit)
Espace d'actions structure : sortie 5x5x52 = 1300, encodant naturellement position x mouvement, avec masquage des actions illegales post-reseau
